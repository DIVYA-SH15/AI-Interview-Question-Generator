```python
from google import genai
from getpass import getpass
```

```python
from getpass import getpass
api_key = getpass("Enter your Gemini API Key: ")
```

```python
client = genai.Client(api_key=api_key)
```

```python

print("AI INTERVIEW QUESTION GENERATOR")
print("-" * 50)
```

```python
job_role = input("Enter the job role: ")
skills = input("Enter required skills (comma separated): ")
experience = input("Enter experience level (Fresher / Intermediate / Experienced): ")
question_type = input("Enter question type (Technical / Behavioral / Mixed): ")
num_questions = input("Enter number of questions: ")
```

```python
prompt = f'''
You are an expert interview preparation assistant.

Generate interview questions based on the following details.

Job Role:
{job_role}

Required Skills:
{skills}

Experience Level:
{experience}

Question Type:
{question_type}

Number of Questions:
{num_questions}

Instructions:
1. Generate clear and relevant interview questions.
2. Match the difficulty with the experience level.
3. Focus on the required skills.
4. Number the questions properly.
5. Keep questions professional and suitable for interviews.
'''
```

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

print("\nGENERATED INTERVIEW QUESTIONS")
print("-" * 50)
print(response.text)
```
